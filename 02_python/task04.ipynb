```python
try:
    # 1つ目の数を入力
    num1 = float(input("1つ目の数を入力してください: "))
    
    # 2つ目の数を入力
    num2 = float(input("2つ目の数を入力してください: "))
    
    result = num1 / num2
    print(f"計算結果: {result}")

# 0で割った場合
except ZeroDivisionError:
    print("エラー: 0 で割ることはできません")

# 数値以外が入力された場合
except ValueError:
    print("エラー: 数値を入力してください")
```
