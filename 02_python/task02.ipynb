```python
price_without_tax = int(input("税抜価格を入力: "))
tax = 10        # 消費税
price_with_tax = price_without_tax * (1+(tax/100))  # 税込価格
shipping = 350  # 送料(税込価格が2000円以上)

if price_with_tax >= 2000:
    print("送料は無料です")
    print("送料込税込価格は", int(price_with_tax), "円です")
elif price_with_tax < 2000:
    print("送料は", shipping, "円です")
    print("送料込税込価格は", int(price_with_tax + shipping), "円です")
```
