```python
# 入力された英単語を格納するリスト
input_words = []

# アルファベットごとの数を格納する辞書
alphabet_count = {
    'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0,
    'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0,
    'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0,
    'v':0, 'w':0, 'x':0, 'y':0, 'z':0
}

# 入力の受付（空文字で終了）
while True:
    word = input("英単語を入力してください：")
    
    if word == "":
        break
    
    input_words.append(word)
    
    # 各文字をカウント
    for char in word:
        if char in alphabet_count:
            alphabet_count[char] += 1

# 英単語を順番に並べ替え
input_words.sort()

# 入力された英単語の表示
print(f"入力した英単語：{input_words}")

# 各アルファベットの出現回数を表示（0より大きいもののみ）
for key in alphabet_count:
    if alphabet_count[key] > 0:
        print(f"{key}が{alphabet_count[key]}個ありました")
```
